```python
import numpy as np
import time
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X_train = np.array([1.4, 2.2, 3.1, 4.7, 5.3, 6.8, 7.2, 8.8, 9.3, 10.7]).reshape(-1,1)
y_train = np.array([18, 28, 38, 57, 58, 67, 78, 92, 109, 126]).reshape(-1,1)
X_test = np.array([3.9, 6]).reshape(-1,1)
```

```python
lr = LinearRegression()
t = time.time()
lr.fit(X_train,y_train)
t = time.time()-t
y_pred = lr.predict(X_test)
print(y_pred)
print('Time Taken To Train Data is : ',t)
```

```python
plt.scatter(X_train,y_train,c='b')
plt.scatter(X_test,y_pred,c='y')
plt.legend(['Test Data','Train Data'])
plt.xlabel('Exprience')
plt.ylabel('Salary In Thousand')
plt.plot(X_train,lr.predict(X_train),c='r')
plt.show()
```
